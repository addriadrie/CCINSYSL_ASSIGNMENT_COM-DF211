1. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | cat | dog | apple | banana | orange


```python
def toLower(input):
  print(input.lower())

toLower(input = input("Enter:"))
```

2. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | CT | DG | PPL | BNN | RNG


```python
def removeVowel(input):
  for x in input:
    if x not in "AEIOUaeiou":
      print(x, end='')

removeVowel(input = input("Enter:"))
```

3. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | TAC | GOD | ELPPA | ANANAB | EGNARO


```python
def reverse(input):
  rev = ""
  for x in input:
    rev = x + rev
  return rev

reverse(input = input("Enter:"))
```

4. Find the function **f(x)**  

| Input  | 2 | 3 | 4 | 5 | 6 |  
| --- | --- | --- | --- | --- | --- |
| Output | 4 | 9 | 16 | 25 | 36


```python
x = 3

def squared(input):
  sq = input * input
  return sq

print(squared(x))
```

5. Find the function **f(x)**  

| Input  | 64 | 27 | 1000 | 729 | 512 |  
| --- | --- | --- | --- | --- | --- |
| Output | 4 | 3 | 10 | 9 | 8

```python
import math

x = 512

def cubeRoot(input):
  cr = int(math.ceil(input**(1/3)))
  return cr

print(cubeRoot(x))
```
